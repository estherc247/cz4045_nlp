```python
# pandas
import pandas as pd

# random
import random

# NLTK packages
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tag import pos_tag as pos_tagger
from nltk import UnigramTagger

# Textblob
from textblob import TextBlob
```

```python
#import the json-file and put it in a dataframe
df = pd.read_json('reviewSelected100.json', lines=True, encoding = "ISO-8859-1")

#dataframe
df
```

```python
#select all the text
text = df['text'].astype('str') 

#join all the text into one string
all_text = ''.join(map(str, text))


#full string
#all_text
```

```python
#separating the string into sentences
sentences = sent_tokenize(all_text)

#sentences
```

```python
#select 5 random sentences
random_5_sentences = random.choices(sentences, k=5)

pd.options.display.max_colwidth = 300
pd.DataFrame(random_5_sentences, columns=['sentence'])
```

```python
def pos_tag(chosen_sentence):
    # using NLTK
    random_sentence_nltk = nltk.pos_tag(word_tokenize(chosen_sentence))

    # using textblob
    random_sentence_blob = TextBlob(chosen_sentence)
    random_sentence_blob.tags

    # respective table
    table_sentence_nltk = pd.DataFrame(random_sentence_nltk, columns=['word', 'POS_Tag'])
    table_sentence_blob = pd.DataFrame(random_sentence_blob.tags, columns=['word', 'POS_Tag'])
                                        
    return table_sentence_nltk, table_sentence_blob
```

```python
# Function for display two tables side by side
from IPython.display import display_html
from itertools import chain,cycle
def display_side_by_side(*args,titles=cycle([''])):
    html_str=''
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
        html_str+='<th style="text-align:center"><td style="vertical-align:top">'
        html_str+=f'<h2>{title}</h2>'
        html_str+=df.to_html().replace('table','table style="display:inline"')
        html_str+='</td></th>'
    display_html(html_str,raw=True)
```

```python
#nltk.download('averaged_perceptron_tagger')
for i in range(len(random_5_sentences)):
    nltk_s, textb_s = pos_tag(random_5_sentences[i])
    print("---------- Sentence "+ str(i+1) + " ----------")
    print("Sentence: " + random_5_sentences[i])
    display_side_by_side(nltk_s, textb_s, titles=['NLTK','TextBlob'])
```

### Observation
Both NLTK and TextBlob produces the same POS_TAG, just that NLTK takes into consideration punctuations as well.

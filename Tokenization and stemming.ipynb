```python
import json
import random

#NLTK packages
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import ngrams
from nltk.stem import PorterStemmer
from nltk import FreqDist
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
from random import randint
from random import choice
```

```python
with open("reviewSelected100.json", 'r') as read_file:
    data = [json.loads(line) for line in read_file]
```

```python
no_of_reviews = len(data)
```

```python
biz_type=[]
for i in range(len(data)):
    biz_id = data[i]['business_id']
    biz_type.append(biz_id)
```

```python
u_biz_id= [] #id list of unique business ids
for x in range(len(biz_type)):
    if biz_type[x] not in u_biz_id:
        count = 0
        for y in range(len(biz_type)):
            if biz_type[y] == biz_type[x]:
                count = count + 1
        u_biz_id.append(biz_type[x])        
        

```

```python
random_no_1 = randint(0,len(u_biz_id))
b1 = u_biz_id[random_no_1] #using seed number generator to find a random business id ,every run will be diff
B1 = {"small_dataset": []}
count = 0
for i in range(len(data)):
    if data[i]['business_id'] == b1:
        B1['small_dataset'].append(data[i])
```

```python
stop_words = set(stopwords.words('english')+ list(ENGLISH_STOP_WORDS))
```

```python
all_words = ''
for i in range(len(B1['small_dataset'])):
    all_words += B1['small_dataset'][i]['text'].lower()
all_tokens =  word_tokenize(all_words)
```

```python
tokens = []
for i in all_tokens:
    if i.isalpha() and i not in stop_words:
        tokens.append(i)
```

```python
token_dist = FreqDist(tokens)
dist = pd.DataFrame(token_dist.most_common(10),columns=['Word', 'Frequency'])

dist
```

```python
#Word frequency after stemming

porter = PorterStemmer()
stem_word =[porter.stem(word) for word in tokens]
stem_word_dist = FreqDist(stem_word)
stem_dist = pd.DataFrame(stem_word_dist.most_common(10),columns=['Word', 'Frequency'])

stem_dist
```

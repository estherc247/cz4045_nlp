## remove all punctuation (can try just ending punctuation)

```python
new_dict = {}
count = 0
for k, v in previous_dict.items():
    new_key = k.translate(str.maketrans('', '', string.punctuation))
    if new_key in new_dict:
        count += 1
        new_dict[new_key] += 1
#         print('have')
    else:
        new_dict[new_key] = v
print(count)
```

```python
dict_sorted_r = {k: v for k, v in sorted(new_dict.items(), key=lambda item: item[1], reverse=True)}
```

```python
dict_sorted_r
```

## x and y --> x, y

```python
and_list = []
for k in dict_sorted_r:
    if k.find(' and ') != -1:
        and_list.append(k)
```

```python
and_list
```

```python
# those len more than 8 -- probably parsing error
and_list_remove_l = [i for i in and_list if len(i.split()) < 8]
```

```python
len(and_list)
```

```python
len(and_list_remove_l)
```

```python
and_list_remove_l
```

```python
new_dict_1 = new_dict.copy()
print('lens of dict before {}'.format(len(new_dict_1)))
count = 0
for and_ in and_list_remove_l:
    split = and_.split(' and ')
    for adjp in split:
        if adjp in new_dict_1:
            new_dict_1[adjp] += 1
            count += 1
        else:
            new_dict_1[adjp] = new_dict_1[and_] 
    del new_dict_1[and_] 
print('lens of dict after {}'.format(len(new_dict_1)))
print(count)
```

```python
dict_sorted_r_a = {k: v for k, v in sorted(new_dict_1.items(), key=lambda item: item[1], reverse=True)}
```

```python
dict_sorted_r_a 
```

```python
count_mt_20 = [k for k, v in dict_sorted_r_a.items() if v >= 20]
len(count_mt_20)
```
